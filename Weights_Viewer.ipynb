```python
import os
import pickle as cPickle
from os.path import isfile, join
import sys

# import theano 
import numpy as np 
import matplotlib.pyplot as plt
plt.style.use('ggplot')

# from CNN.CNN import *
from helpers.utilities import SaveData
```

```python
def readFile(file_name):
    if(os.path.isfile(file_name)):
        f = open(file_name, 'rb')
        params = cPickle.load(f)
        f.close()
    else:
        print "params not founded"
        sys.exit(0)
    return params       
```

```python
params_path = 'weights/FCNN-comp1'
params = [50,1]

param_toRead = join(params_path, ('%i.%i.pkl' %(params[0],params[1])))
currentP = readFile(param_toRead)
ArchiveP = readFile('Archive.pkl')

# print currentP[0].flatten()[0]
# print ArchiveP[0].flatten()[0]
for i in range(0,len(currentP),1):
    B = currentP[i].flatten()
    BP = ArchiveP[i].flatten()
    BUR = BP-B
#     print float(BUR.sum())

    plt.figure(figsize=(10,10))
    plt.subplot(131)
    plt.plot(B, 'g.')
    plt.subplot(132)
    plt.plot(BP, 'm.')
    plt.subplot(133)
    plt.plot(BUR,'r.')
    # plt.suptitle('B7')
    # plt.hist(B7_reshaped, bins=10)
    plt.show() 

SaveData(currentP, 'Archive.pkl')
```
